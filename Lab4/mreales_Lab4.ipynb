## SymPy’s polynomials 
- https://docs.sympy.org/latest/modules/polys/basics.html#polynomials 

$$ P_n(x) =  a_n x^n + a_{n-1} x^{n-1} + a_{n-2} x^{n-2} +\cdots +  a_1 x  + a_0$$

$$P_n(x) = 0 $$

$$a_n x^n + a_{n-1} x^{n-1} + a_{n-2} x^{n-2} +\cdots +  a_1 x  + a_0 = 0$$

$$(x-r_1) Q_{n-1}(x) = P_n(x)$$
$$(x-r_1) (x - r_2) Q_{n-2}(x) = P_n(x)$$
$$ \vdots $$

- x^10 - 71 x^9 + 576 x^8 - 1116 x^7 + x^3 - 71 x^2 + 576 x - 1116 = 
(x + 1) (x - 3) (x - 6) (x - 62) (x^6 - x^5 + x^4 - x^3 + x^2 - x + 1)

```
from sympy import Symbol
from sympy import div

x = Symbol('x')

p = x**10 - 71*x**9 + 576*x**8 - 1116*x**7 + x**3 - 71*x**2 + 576*x - 1116

p, r = div(p,  x-1)

print(p)
print(r)

p, r = div(p,  x-2)

print(p)
print(r)

p, r = div(p,  x-3)

print(p)
print(r)

p, r = div(p,  x-4)

print(p)
print(r)

```

```
from sympy import Symbol, solveset, Eq
x = Symbol('x')
solveset(Eq(x**10 - 71*x**9 + 576*x**8 - 1116*x**7 + x**3 - 71*x**2 + 576*x - 1116, 0), x)
```

```
import sympy as sp

# Definir la variable simbólica x
x = sp.symbols('x')
# Definir la ecuación cuadrática
eq = x**10 - 71*x**9 + 576*x**8 - 1116*x**7 + x**3 - 71*x**2 + 576*x - 1116
# Encontrar las soluciones de la ecuación cuadrática
sol = sp.solveset(eq, x)
# Imprimir las soluciones
print("Las soluciones de la ecuación son: ", sol)
```

## SymPy’s Symbolic and Numercical Complex Evaluations


image.png

```
from sympy import symbols, I, N, pi, expand
x1, x2, y1, y2, z1, z2 = symbols("x1 x2 y1 y2 z1 z2", real=True)  
x = 2+3*I
y = -4-2*I
z = 1+2*I

print(x*y*z)
print(expand(x*y*z))
print(expand((x*y)*z))
print(expand(x*(y*z)))
w=1/(pi + I)
print('w=',w)
w=N(w)
print('w=',w)
```
